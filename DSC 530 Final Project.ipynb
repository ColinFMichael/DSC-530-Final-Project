```python
def EvalCdf(sample, x):
    count = 0.0
    for value in sample:
        if value <= x:
            count += 1

    prob = count / len(sample)
    return prob
```

```python
cdf = thinkstats2.Cdf(comedy.rating, label='Comedy')
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Comedy', ylabel='CDF', loc='upper left')
```

```python
comedyholder=comedy.rating

log_comedyholder = np.log10(comedyholder)

cdf = thinkstats2.Cdf(comedyholder, label="data")
cdf_log = thinkstats2.Cdf(log_comedyholder, label="data")

thinkplot.PrePlot(cols=2)

mu, sigma = log_comedyholder.mean(), log_comedyholder.std()
xs, ps = thinkstats2.RenderNormalCdf(mu, sigma, low=0, high=1)
thinkplot.Plot(xs, ps, label="model", color="0.8")

thinkplot.Cdf(cdf_log)
thinkplot.Config(xlabel="log10 comedy_rating", ylabel="CDF", loc="lower right")
```

```python
def SampleRows(df, nrows, replace=False):
    indices = np.random.choice(df.index, nrows, replace=replace)
    sample = df.loc[indices]
    return sample
```

```python
sample = SampleRows(df, 1000)
length, rating, votes = sample.length, sample.rating, sample.votes
```

```python
thinkplot.Scatter(length, rating, alpha=1)
thinkplot.Config(xlabel='length (minutes)',
                 ylabel='rating',
                 axis=[5, 250, 0, 10],
                 legend=False)

np.corrcoef(length, rating)
```

```python
thinkplot.Scatter(votes, rating, alpha=1)
thinkplot.Config(xlabel='votes',
                 ylabel='rating',
                 axis=[1, 50000, 0, 10],
                 legend=False)

np.corrcoef(votes, rating)
```

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
cleaned = df.dropna(subset=['length', 'rating'])
data = cleaned.length.values, cleaned.rating.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
pvalue
```
